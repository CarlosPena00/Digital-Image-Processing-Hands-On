## Introduction To OpenCV in Python

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Load Image

```python
lenna_path = r"../data/Lenna.png"
lenna_img  = cv2.imread(lenna_path)
gray_lena   = cv2.cvtColor(lenna_img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 10))
plt.subplot(131)
plt.title("BGR Image")
plt.imshow(lenna_img)
plt.subplot(132)
plt.title("RGB Image")
plt.imshow(lenna_img[..., ::-1])
plt.subplot(133)
plt.title("Gray Image")
plt.imshow(gray_lena, cmap='gray')

plt.show()
```

## Plot

```python
black  = np.zeros([100, 100], dtype=np.uint8)
gray   = np.ones([100, 100], dtype=np.uint8) * 100
white  = np.ones([100, 100], dtype=np.uint8) * 255
plt.figure(figsize=(5, 5))
plt.subplot(231);plt.title("Black");plt.imshow(black)
plt.subplot(232);plt.title("Gray");plt.imshow(gray)
plt.subplot(233);plt.title("White");plt.imshow(white)
plt.subplot(234);plt.title("Black");plt.imshow(black, vmax=255, vmin=0)
plt.subplot(235);plt.title("Gray");plt.imshow(gray, vmax=255, vmin=0)
plt.subplot(236);plt.title("White");plt.imshow(white, vmax=255, vmin=0)
## Note: Matplotlib normalize the input image (if one channel)
```

## Create 2.8 Figure

```python
black  = np.zeros([100, 100], dtype=np.uint8)
gray   = np.ones([100, 100], dtype=np.uint8) * 100
white  = np.ones([100, 100], dtype=np.uint8) * 255
black[33:66,33:66] = 127
gray[33:66,33:66] = 127
white[33:66,33:66] = 127

plt.figure(figsize=(10, 10))
plt.subplot(131);plt.title("Black");plt.imshow(black, vmax=255, vmin=0, cmap='gray')
plt.subplot(132);plt.title("Gray");plt.imshow(gray, vmax=255, vmin=0, cmap='gray')
plt.subplot(133);plt.title("White");plt.imshow(white, vmax=255, vmin=0, cmap='gray')
```

## Draw 
#### Numpy vs OpenCV

```python
## Numpy
src  = np.zeros([50, 50, 3], dtype=np.uint8)
src[10:20, 30:40] = [100, 0, 0]
src[10:12, 30:32] = [255, 0, 0]

## OpenCV
pt1 = (10, 20)
pt2 = (30, 40)
cv2.rectangle(src, pt1, pt2, (0, 100, 0), thickness=-1)
cv2.line(src, pt1, pt1, (0, 255, 0))
cv2.line(src, pt2, pt2, (0, 255, 0))

plt.figure(figsize=(5, 5))
plt.imshow(src, vmax=255, vmin=0, cmap='gray')
```

## OpenCV Draw

```python
src  = np.zeros([256, 256, 3], dtype=np.uint8)
center = (50, 200)
cv2.circle(src, center, 5, [0,255,0], 4)
cv2.rectangle(src, (10, 20), (30, 40), (0, 255, 255), 3)
cv2.line(src, (150, 150), (200, 200), (255, 0, 0), 3)
cv2.putText(src, "Alfa", (20, 170), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 3)

plt.figure(figsize=(5, 5))
plt.imshow(src, vmax=255, vmin=0, cmap='gray')
```
